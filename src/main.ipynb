## Task 1 - Data Preparation and Exploration (EDA)

### 1. Load customer_churn.csv into Python.

```python
# Numerical computing libraries

import pandas as pd
import numpy as np

# Visalization libraries

import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline

raw_data = pd.read_csv('../data/customer_churn.csv')

print('=== Dataset Structure ===')
print(f'Shape: {raw_data.shape}')
print('\nColumns:')
print(raw_data.columns.tolist())

print('\nData Types and Non-Null Counts:')
raw_data.info()

print('\nMissing Values Per Column:')
print(raw_data.isnull().sum())

print('\n=== Summary Statistics ===')
display(raw_data.describe(include='all'))

print('\n=== Identify and Handle Missing Values ===')
missing_counts = raw_data.isnull().sum()
total_missing = int(missing_counts.sum())
print('Total missing values in dataset:', total_missing)
display(missing_counts[missing_counts > 0])

clean_data = raw_data.copy()
if total_missing > 0:
    numeric_cols = clean_data.select_dtypes(include=['number']).columns
    categorical_cols = clean_data.select_dtypes(exclude=['number']).columns

    for col in numeric_cols:
        clean_data[col] = clean_data[col].fillna(clean_data[col].median())

    for col in categorical_cols:
        clean_data[col] = clean_data[col].fillna(clean_data[col].mode().iloc[0])

    print('Missing values handled: numeric -> median, categorical -> mode.')
else:
    print('No missing values found. No handling needed.')

print('Remaining missing values after handling:', int(clean_data.isnull().sum().sum()))

print('\n=== Exploratory Data Analysis (EDA) ===')
eda_data = clean_data.copy()
numeric_cols = eda_data.select_dtypes(include=['number']).columns.tolist()
numeric_cols_for_eda = [col for col in numeric_cols if col != 'CustomerID']

print('Numeric columns used for EDA:', numeric_cols_for_eda)

# 1) Histograms
sns.set_style('whitegrid')
n_features = len(numeric_cols_for_eda)
n_cols = 3
n_rows = int(np.ceil(n_features / n_cols))
fig, axes = plt.subplots(n_rows, n_cols, figsize=(18, 4.5 * n_rows))
axes = np.array(axes).reshape(-1)
palette = sns.color_palette('Set2', n_colors=n_features)

for i, col in enumerate(numeric_cols_for_eda):
    ax = axes[i]
    series = eda_data[col].dropna()
    ax.hist(series, bins=20, color=palette[i], edgecolor='white', alpha=0.9)
    ax.axvline(series.mean(), color='black', linestyle='--', linewidth=1.5, label='Mean')
    ax.axvline(series.median(), color='firebrick', linestyle=':', linewidth=1.8, label='Median')
    ax.set_title(f'{col} Distribution', fontsize=11, pad=8)
    ax.set_xlabel(col)
    ax.set_ylabel('Frequency')
    ax.legend(fontsize=8)

for j in range(n_features, len(axes)):
    axes[j].set_visible(False)

fig.suptitle('Histograms of Numeric Features', fontsize=15, y=1.01)
fig.tight_layout()
plt.show()

# 2) Box plots
plt.figure(figsize=(14, 7))
boxplot_series = [eda_data[col].dropna() for col in numeric_cols_for_eda]
plt.boxplot(boxplot_series, vert=False, tick_labels=numeric_cols_for_eda, patch_artist=True)
plt.title('Box Plots of Numeric Features')
plt.xlabel('Value')
plt.ylabel('Feature')
plt.tight_layout()
plt.show()

# 3) Correlation analysis
numeric_df = eda_data.loc[:, numeric_cols_for_eda].copy()
corr_matrix = pd.DataFrame(numeric_df).corr(method='pearson')
plt.figure(figsize=(10, 8))
sns.heatmap(corr_matrix, annot=True, fmt='.2f', cmap='coolwarm', square=True)
plt.title('Correlation Heatmap')
plt.tight_layout()
plt.show()

if 'Churn' in corr_matrix.columns:
    print('\nCorrelation with Churn (highest to lowest):')
    churn_corr = corr_matrix.loc[:, 'Churn']
    print(churn_corr.sort_values(ascending=False, kind='quicksort'))

# 4) Categorical feature EDA
print('\n=== Categorical EDA ===')
categorical_cols = eda_data.select_dtypes(include=['object', 'string', 'category']).columns.tolist()
print('Categorical columns used for EDA:', categorical_cols)

for col in categorical_cols:
    print(f'\n--- {col} ---')

    # Frequency and percentage tables
    counts = eda_data[col].value_counts(dropna=False)
    percents = eda_data[col].value_counts(dropna=False, normalize=True).mul(100).round(2)
    summary_table = pd.DataFrame({'count': counts, 'percent': percents})
    display(summary_table)

    # Count plot (bar chart)
    plt.figure(figsize=(7, 4))
    counts.plot(kind='bar', color='steelblue', edgecolor='black')
    plt.title(f'{col} - Category Counts')
    plt.xlabel(col)
    plt.ylabel('Count')
    plt.xticks(rotation=30, ha='right')
    plt.tight_layout()
    plt.show()

    if 'Churn' in eda_data.columns:
        # Contingency table and normalized contingency table
        print(f'Contingency table: {col} vs Churn')
        ct = pd.crosstab(eda_data[col], eda_data['Churn'])
        display(ct)
        print(f'Normalized contingency table (row %): {col} vs Churn')
        display(pd.crosstab(eda_data[col], eda_data['Churn'], normalize='index').round(3))

        # Churn rate by category
        churn_rate = pd.Series(eda_data.groupby(col)['Churn'].mean()).sort_values().iloc[::-1]
        print(f'Churn rate by {col}:')
        display(churn_rate.round(3))

        plt.figure(figsize=(7, 4))
        churn_rate.mul(100).plot(kind='bar', color='indianred', edgecolor='black')
        plt.title(f'Churn Rate by {col}')
        plt.xlabel(col)
        plt.ylabel('Churn Rate (%)')
        plt.xticks(rotation=30, ha='right')
        plt.ylim(0, 100)
        plt.tight_layout()
        plt.show()

```
