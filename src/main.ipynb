## Task 1 - Data Preparation and Exploration (EDA)

### 1. Load customer_churn.csv into Python.

```python
# Numerical computing libraries

import pandas as pd
import numpy as np

# Visalization libraries

import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline

raw_data = pd.read_csv('../data/customer_churn.csv')

print('=== Dataset Structure ===')
print(f'Shape: {raw_data.shape}')
print('\nColumns:')
print(raw_data.columns.tolist())

print('\nData Types and Non-Null Counts:')
raw_data.info()

print('\nMissing Values Per Column:')
print(raw_data.isnull().sum())

print('\n=== Summary Statistics ===')
display(raw_data.describe(include='all'))

print('\n=== Identify and Handle Missing Values ===')
missing_counts = raw_data.isnull().sum()
total_missing = int(missing_counts.sum())
print('Total missing values in dataset:', total_missing)
display(missing_counts[missing_counts > 0])

clean_data = raw_data.copy()
if total_missing > 0:
    numeric_cols = clean_data.select_dtypes(include=['number']).columns
    categorical_cols = clean_data.select_dtypes(exclude=['number']).columns

    for col in numeric_cols:
        clean_data[col] = clean_data[col].fillna(clean_data[col].median())

    for col in categorical_cols:
        clean_data[col] = clean_data[col].fillna(clean_data[col].mode().iloc[0])

    print('Missing values handled: numeric -> median, categorical -> mode.')
else:
    print('No missing values found. No handling needed.')

print('Remaining missing values after handling:', int(clean_data.isnull().sum().sum()))

print('\n=== Exploratory Data Analysis (EDA) ===')
eda_data = clean_data.copy()
numeric_cols = eda_data.select_dtypes(include=['number']).columns.tolist()
numeric_cols_for_eda = [col for col in numeric_cols if col != 'CustomerID']

print('Numeric columns used for EDA:', numeric_cols_for_eda)

# 1) Histograms
eda_data[numeric_cols_for_eda].hist(bins=20, figsize=(16, 10), edgecolor='black')
plt.suptitle('Histograms of Numeric Features', y=1.02)
plt.tight_layout()
plt.show()

# 2) Box plots
plt.figure(figsize=(14, 7))
boxplot_data = eda_data[numeric_cols_for_eda].melt(var_name='Feature', value_name='Value')
sns.boxplot(data=boxplot_data, x='Value', y='Feature')
plt.title('Box Plots of Numeric Features')
plt.tight_layout()
plt.show()

# 3) Correlation analysis
numeric_df = eda_data.loc[:, numeric_cols_for_eda].copy()
corr_matrix = pd.DataFrame(numeric_df).corr(method='pearson')
plt.figure(figsize=(10, 8))
sns.heatmap(corr_matrix, annot=True, fmt='.2f', cmap='coolwarm', square=True)
plt.title('Correlation Heatmap')
plt.tight_layout()
plt.show()

if 'Churn' in corr_matrix.columns:
    print('\nCorrelation with Churn (highest to lowest):')
    churn_corr = corr_matrix.loc[:, 'Churn']
    print(churn_corr.sort_values(ascending=False, kind='quicksort'))

```
